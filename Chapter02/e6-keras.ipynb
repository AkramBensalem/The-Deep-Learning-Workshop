# Classify structured data

## Import TensorFlow and other libraries

```python
import tensorflow as tf
import pandas as pd 
 

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##Import Keras libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
dataframe = pd.read_csv('data.csv')
dataframe.head()
```

```python
plt.scatter(dataframe.x1,dataframe.x2,c=dataframe.label)
```

```python

x_input=dataframe[['x1','x2']].values
y_label=dataframe[['label']].values

 
```

```python
mymodel = Sequential()
```

```python
mymodel.add(Dense(units = 1,input_dim=2, activation = 'sigmoid'))
```

```python
## Compliling
mymodel.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
mymodel.summary()
```

```python
mymodel.fit(x_input, y_label, epochs=1000)
```
