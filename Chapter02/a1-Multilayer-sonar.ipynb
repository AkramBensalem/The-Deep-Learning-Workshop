```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
##Import Keras libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
data=pd.read_csv('sonar.csv')
data.head()
```

```python
X_input=data.drop(labels='Class', axis=1) 
Y_label=data[['Class']].values
```

```python
X_input.head()
```

```python
labelencoder_Y = LabelEncoder()
 
Y_label=labelencoder_Y.fit_transform(Y_label)
```

```python
Y_label=Y_label.reshape([208,1])
```

```python
Y_label
```

```python
mymodel = Sequential()
```

```python
mymodel.add(Dense(300,input_dim=60, activation = 'relu'))
mymodel.add(Dense(200,  activation = 'relu'))
mymodel.add(Dense(100,  activation = 'relu'))
mymodel.add(Dense(1,  activation = 'sigmoid'))
```

```python
mymodel.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
mymodel.fit(X_input, Y_label, epochs=20)
```
